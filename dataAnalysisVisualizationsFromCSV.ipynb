```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
avo=pd.read_csv('Avocado.csv')
```

```python
avo.head()
```

```python

avo.sample(25)
```

```python
avo['Date'] = pd.to_datetime(avo['Date'],
                            errors ='coerce')
avo['weekNumber'] = avo['Date'].dt.week
```

```python
avo.head(3)
```

```python
avo = avo.drop(['ID','Date'], axis=1)
```

```python
avo.head(2)
```

```python
avo['type']=avo['type'].map({'conventional':1,'organic':0})
```

```python
avo.tail(2)
```

```python
plt.subplots(figsize=(10,10))
sns.boxplot(x='year', y='Average Price', data=avo)
plt.savefig('Avo_boxplots.png')
```

```python
fig=plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.scatter(avo['Average Price'],avo['Total Volume'])
plt.title("Price vs Volume")
plt.ylabel("Total Volume")
plt.xlabel("Average Price")
plt.legend(loc='best')


plt.subplot(1,2,2)
plt.scatter(avo['Average Price'], avo["Total Bags"])
plt.title("Average Price vs Total bags")
plt.ylabel("Total Bags")
plt.xlabel("Average Price")
plt.legend(loc='best')

plt.savefig('Avo_regplots.png')
plt.show()

```

```python
avo['Average Price'].corr(avo['Total Volume'])
```

```python
avo['Average Price'].corr(avo['Total Bags'])
```

```python
corr = avo.corr()
high_corr = corr[((corr>=.7) | (corr<= -.7))] # & (corr!=1.0)]
plt.figure(figsize=(12,8))
sns.heatmap(high_corr, cmap="Greens")
```

```python
fig=plt.figure(figsize=(20,25))

plt.subplot(3,2,1)
plt.bar(avo['year'],avo['Total Bags'], align='center')
plt.title('Total Bags_yearly')
plt.ylabel('Number of Avocados Bags sold')
plt.xlabel('year')

plt.subplot(3,2,2)
plt.bar(avo['year'],avo['Total Volume'], align='center')
plt.title('Total volume_yearly')
plt.ylabel('Number of Avocados sold')
plt.xlabel('year')

plt.savefig('Avocados.jpg')
plt.show()
```
